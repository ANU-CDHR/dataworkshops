2.1.2 Using BeautifulSoup python library to get data
------
**Get data from Wikipedia page table via url request. Get list of explorers from https://en.wikipedia.org/wiki/List_of_explorers which contain explorer names, nationalities, centuries and main areas explored.**

```python
# import BeautifulSoup and other reqired libraries
from bs4 import BeautifulSoup
import requests
import os
import codecs
```

```python
# set Wikipedia page url to be scraped
url="https://en.wikipedia.org/wiki/List_of_explorers"
#url="https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(PPP)_per_capita"
```

**Get page via url request.**

```python
# access page via url request
wikipage = requests.get(url)
```

**Create BeautifulSoup object with page content.**

```python
# create BeautifulSoup object with page content
soup = BeautifulSoup(wikipage.content)
```

**Find all tables in the page.**

```python
# find all tables in the page
tables = soup.findAll("table", {"class": "wikitable"})
```

**Show table.**

```python
# show tables
for i, table in enumerate(tables):
    # print table number
    print("#"*10 + "Table {}".format(i) + '#'*10)
    # print first 50 text
    print(table.text[:50])
    # print divider
    print('.'*80)
print("#"*80)
```

**Create function to export TSVs (tab-separated values file) from Wiki page tables.** 

```python
# export CSVs from Wiki page tables
def exportWikiTB(Url, tables):
    for tn, table in enumerate(tables):

        # preinit list of lists
        rows = table.findAll("tr")
        row_lengths = [len(r.findAll(['th', 'td'])) for r in rows]
        ncols = max(row_lengths)
        nrows = len(rows)
        data = []
        for i in range(nrows):
            rowD = []
            for j in range(ncols):
                rowD.append('')
            data.append(rowD)

        # process html
        for i in range(len(rows)):
            row = rows[i]
            rowD = []
            cells = row.findAll(["td", "th"])
            for j in range(len(cells)):
                cell = cells[j]

                #lots of cells span cols and rows so lets deal with that
                cspan = int(cell.get('colspan', 1))
                rspan = int(cell.get('rowspan', 1))
                l = 0
                for k in range(rspan):
                    # Shifts to the first empty cell of this row
                    while data[i + k][j + l]:
                        l += 1
                    for m in range(cspan):
                        cell_n = j + l + m
                        row_n = i + k
                        # in some cases the colspan can overflow the table, in those cases just get the last item
                        cell_n = min(cell_n, len(data[row_n])-1)
                        data[row_n][cell_n] += cell.text
                        #print(cell.text)

            data.append(rowD)

        # write data out to tab seperated format
        page = os.path.split(url)[1]
        fname = 'output_{}_t{}.tsv'.format(page, tn)
        f = codecs.open(fname, 'w')
        for i in range(nrows):
            rowStr = '\t'.join(data[i])
            rowStr = rowStr.replace('\n', '')
            #print(rowStr)
            f.write(rowStr + '\n')

        f.close()
```

**Run the function to export TSV file.**

```python
exportWikiTB(url, tables)
```

**Combine all above and create the function to plot wordcloud from Wikipeida url.**

```python

from bs4 import BeautifulSoup
import requests
import os
import re
import codecs
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import nltk
#nltk.download()
from nltk.corpus import stopwords
#import spacy 


# define function to plot wordcloud from Wikipeida url
def wiki_cloud(url,lang="english", **kwargs):
    # access page via url request
    wikipage = requests.get(url)
    # create BeautifulSoup object with page content
    soup = BeautifulSoup(wikipage.text)
    # select paragraphs
    paragraphs = soup.select('p')
    paragraphs = [paragraph.text for paragraph in paragraphs]
    # remove [] citation sybmols from text
    paragraphs = [re.sub(r'\[\d+\]','',paragraph.strip())  for paragraph in paragraphs if paragraph.strip() != '']
    # clean page content remove symbols and changing line
    content = re.sub(r'==.*?==+', '', " ".join(paragraphs))
    content = content.replace('\n', '')
    #print(content)
    #use different language stopwords
    nltk_sw = stopwords.words(lang)
    #nlp = spacy.load('es_core_web_sm')


    # remove stopwords and generate wordcloud content
    wordcloud = WordCloud(stopwords=nltk_sw, **kwargs).generate(content)
    # show wordcloud
    plt.figure(figsize = (10,10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
url="https://en.wikipedia.org/wiki/Democracy"
wiki_cloud(url,'english')
```

**Find other language versions of the same Wikipedia page under the languages menu on the left side of the page.**

```python
#Spanish version of the Wikipedia page
url="https://es.wikipedia.org/wiki/Democracia"
wiki_cloud(url, 'spanish')
```

```python

#Chinese version of the Wikipedia page
url="https://zh.wikipedia.org/wiki/%E6%B0%91%E4%B8%BB"
#Use different font for the Chinese version of the word cloud
font_filename = 'fonts/STFangSong.ttf'
wiki_cloud(url, 'chinese',font_path=font_filename)
```
